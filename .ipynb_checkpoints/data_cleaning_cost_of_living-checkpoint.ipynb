```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Define the list of city names and locate csv files
cities_list = ["Atlanta","Boston","Chicago","Houston","Los Angeles","New York",
              "Philadelphia","Seattle","San Francisco","Washington DC"]
folder = "./Resources/cost_of_living_data/"

# Create empty dataframe with 66 columns
compiled_df = pd.DataFrame(columns=range(1,67))
compiled_df   
```

```python
# Iterate through the list of city names and read corresponding csv file
for city in cities_list:
    csv = f"./{folder}/{city}.csv"
    #print (csv)
    df = pd.read_csv(csv,encoding="utf-8",header=None)
    df_t = df.transpose()
    
    df_t['City'] = city
    df_t = df_t.set_index('City')
    df_t = df_t.tail(-1)

    df_median = df_t.head(1)
    df_range = df_t.tail(1)
    
    #append
    compiled_df = compiled_df.append(df_median)
```

```python
compiled_df
```

```python
# Remove irrelevant columns
compiled_new_df = compiled_df.drop([0,9,29,38,42,46,49,54,59,62,65,66,67], axis=1)
compiled_new_df
```

```python
# Relabel column numbers
compiled_new_df.columns = [np.arange(0,compiled_new_df.shape[1])]

# Remove $ signs and empty spaces
for col in np.arange(0,54): #cols
    for row in np.arange(0,10): #rows
        compiled_new_df.iloc[row][col] = compiled_new_df.iloc[row][col].str.replace(u'\xa0', u'')
        compiled_new_df.iloc[row][col] = compiled_new_df.iloc[row][col].str.replace(u'$', u'')
        compiled_new_df.iloc[row][col] = compiled_new_df.iloc[row][col].str.replace(u',', u'')
         
# Divide the International Primary School, Yearly value by 12 to determine monthly value
compiled_new_df[42] = [float(x) / 12 for x in compiled_new_df[42].values]
```
