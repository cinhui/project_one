```python
keyword = "Accountants"
location_city = "Chicago, IL"
location_state = "IL"
```

```python
# search job postings for keyword
jobsearch_url = ('https://api.careeronestop.org/v1/jobsearch/'
                '{0}/{1}/{2}/25/0/0/0/100/30').format(user_id,keyword,location_city)
print(jobsearch_url)
jobsearch_response = requests.get(jobsearch_url,
                        headers={'Content-Type':'application/json',
                        'Authorization': 'Bearer {}'.format(api_key)})
print(jobsearch_response.status_code)

# get occupation details on keyword 
occupation_url = ('https://api.careeronestop.org/v1/occupation/'
            '{0}/{1}/{2}/').format(user_id,keyword,location_city)
print(occupation_url)
occupation_response = requests.get(occupation_url,
                headers={'Content-Type':'application/json',
               'Authorization': 'Bearer {}'.format(api_key)})
print(occupation_response.status_code)

# get occupation reports for fast growing occupation
occupationsreport_url = ('https://api.careeronestop.org/v1/occupationsreports/'
                         '{0}/FG/{1}/0/Median/DESC/0/1000/').format(user_id,location_state)
print(occupationsreport_url)
occupationreport_response = requests.get(occupationsreport_url,
                        headers={'Content-Type':'application/json',
                        'Authorization': 'Bearer {}'.format(api_key)})
print(occupationreport_response.status_code)
```
